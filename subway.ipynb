```python
import csv

f = open('subway.csv')
data = csv.reader(f)
mx = 0 #rate의 최댓값 저장
rate = 0 #각 역의 비율 데이터 저장
mx_st = ''

next(data)

for row in data:
  for i in range(4,8):
    row[i] = row[i].replace(',', '')
    row[i] = int(row[i])
  if row[6] != 0 and (row[4]+row[6]) > 100000:
    rate = row[4] / (row[4]+row[6])
    if rate > mx:
      mx = rate
      mx_st = row[3] + '({})'.format(row[1])
print(mx_st, round(mx*100, 2))
```

```python
import csv
import matplotlib.pyplot as plt

f = open("time.csv")
data = csv.reader(f)
next(data)
next(data)

mx = [0]*24
mx_st = ['']*24

for row in data:
    row[3:] = map(int, row[3:])
    for j in range(24):
        a=row[j*2+3]
        if a > mx[j]:
            mx[j] = a
            mx_st[j] = row[2] + ':' + str((j+3)%24) + '시'

plt.rc('font', family='NanumBarunGothic')
plt.bar(range(24), mx)
plt.title('지하철 시간대별 승차 인원')
plt.xticks(range(24), mx_st, rotation=90)
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
f = open("time.csv")
data = csv.reader(f)
next(data)
next(data)

s_in=[0] * 24
s_out=[0] * 24

for row in data:
    row[3:] = map(int, row[3:])
    for i in range(24):
        s_in[i] += row[i*2+3]
        s_out[i] += row[i*2+4]

plt.rc('font', family='NanumBarunGothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(3, 27))
plt.show()
```

```python
from flask import Flask, render_template

app = Flask(__name__, static_folder='static')

@app.route('/')
def index():
    import csv
    import numpy as np
    import matplotlib.pyplot as plt
    f = open("time.csv", encoding="utf-8")
    data = csv.reader(f)
    next(data)

    s_in=[0] * 24
    s_out=[0] * 24

    for row in data:
        row[3:] = map(int, row[3:])
        for i in range(24):
            s_in[i] += row[i*2+3]
            s_out[i] += row[i*2+4]

    # plt.rc('font', family='NanumBarunGothic')
    plt.title('지하철 시간대별 승하차 인원 추이')
    plt.plot(s_in, label='승차')
    plt.plot(s_out, label='하차')
    plt.legend()
    plt.xticks(range(24), range(3, 27))
    plt.savefig('static/graph.png')
    plt.close()  # 이미지 생성 후에는 plt를 닫아주어야 합니다.

    return render_template('/index.html', graph_url='static/graph.png')

@app.route('/riding')
def riding():
    import csv
    import matplotlib.pyplot as plt

    f = open("time.csv", encoding="utf-8")
    data = csv.reader(f)
    next(data)
    next(data)

    mx = [0] * 24
    mx_st = [''] * 24

    for row in data:
        row[3:] = map(int, row[3:])
        for j in range(24):
            a = row[j*2+3]
            if a > mx[j]:
                mx[j] = a
                mx_st[j] = row[2] + ':' + str((j+3) % 24) + '시'

    # plt.rc('font', family='NanumBarunGothic')
    plt.bar(range(24), mx)
    plt.title('지하철 시간대별 승차 인원')
    plt.xticks(range(24), mx_st, rotation=90)
    plt.savefig('static/riding.png')
    plt.close()  # 이미지 생성 후에는 plt를 닫아주어야 합니다.

    return render_template('riding.html', graph_url='static/riding.png')


@app.route('/noriding')
def noriding():
    import csv
    import matplotlib.pyplot as plt

    f = open("time.csv", encoding="utf-8")
    data = csv.reader(f)
    next(data)

    mx = [0] * 24
    mx_st = [''] * 24

    for row in data:
        row[3:] = map(int, row[3:])
        for j in range(24):
            a = row[j*2+4]
            if a > mx[j]:
                mx[j] = a
                mx_st[j] = row[2] + ':' + str((j+3) % 24) + '시'

    # plt.rc('font', family='NanumBarunGothic')
    plt.bar(range(24), mx)
    plt.title('지하철 시간대별 하차 인원')
    plt.xticks(range(24), mx_st, rotation=90)
    plt.savefig('static/noriding.png')
    plt.close()  # 이미지 생성 후에는 plt를 닫아주어야 합니다.

    return render_template('noriding.html', graph_url='static/noriding.png')

if __name__ == '__main__':
    app.run('localhost', port=5023, debug=False)
```
